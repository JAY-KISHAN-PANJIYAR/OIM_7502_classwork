#### 6. Use Python to determine how many days until New Year's Day

```python
import datetime

def days_until_new_year():
    """
    Calculates and prints the number of days until New Year's Day.
    """
    # Get today's date
    today = datetime.date.today()

    # Determine the next New Year's Day. If today is already in the new year,
    # it calculates for the next year.
    new_year_date = datetime.date(today.year + 1, 1, 1)

    # Calculate the difference between the two dates
    difference = new_year_date - today

    # Print the result
    print(f"There are {difference.days} days until New Year's Day.")

# Call the function to run the code
days_until_new_year()
```

#### 7. Write a short program that displays the future value of 1,000 earning 5%  at the end of each year for the next 5 years

```python
# Define the initial investment amount and the interest rate.
principal = 1000
interest_rate = 0.05
num_years = 5

print("Future Value of $1,000 at 5% Interest:")

# Use a loop to calculate the value at the end of each year.
for year in range(1, num_years + 1):
    # Calculate the future value for the current year.
    future_value = principal * (1 + interest_rate)

    # Update the principal for the next year's calculation.
    principal = future_value

    # Display the result for the current year, formatted to two decimal places.
    print(f"End of Year {year}: ${future_value:.2f}")
```

#### 8. Write a short program that prompts a user to enter a stock symbol and press enter. Add each symbol entered to a variable. The program should run until the user presses the enter key without entering anything (empty string) 

```python
"""
This program prompts the user to enter stock symbols and stores them in a list.
The program continues to run until the user presses Enter without entering
a stock symbol.
"""

def collect_stock_symbols():
    """
    Collects stock symbols from user input until an empty string is entered.

    Returns:
        list: A list of the entered stock symbols.
    """
    stocks = []
    print("Enter a stock symbol. Press Enter on an empty line to finish.")

    while True:
        symbol = input("Enter a stock symbol: ").strip().upper()
        # Check if the user entered an empty string
        if not symbol:
            break
        stocks.append(symbol)

    return stocks

if __name__ == "__main__":
    symbols_list = collect_stock_symbols()
    print("\n--- Summary ---")
    if symbols_list:
        print("You entered the following stock symbols:")
        for stock in symbols_list:
            print(f"- {stock}")
    else:
        print("No stock symbols were entered.")
    print("---------------")
```

#### 9. Write the code necessary to calculate the sum of the square differences of a group of values : 
58, 32, 37, 41, 36, 36, 54, 37, 25, 53
##### $\Sigma $(${x}$ - $\bar{x}$)$^2$

```python
"""
Calculates the sum of the square differences for a given dataset.

The sum of square differences is a fundamental statistical calculation,
represented by the formula $\Sigma(x - \bar{x})^2$, where:
- $\Sigma$ is the summation symbol.
- $x$ represents each value in the dataset.
- $\bar{x}$ (x-bar) is the mean (average) of the dataset.

This program calculates this value for the specific dataset provided in the problem.
"""

# The group of values from the problem.
values = [58, 32, 37, 41, 36, 36, 54, 37, 25, 53]

# Step 1: Calculate the mean of the values.
# The mean is the sum of all values divided by the number of values.
number_of_values = len(values)
sum_of_values = sum(values)
mean = sum_of_values / number_of_values

print(f"The values are: {values}")
print(f"Number of values: {number_of_values}")
print(f"Sum of values: {sum_of_values}")
print(f"The mean ($\bar{{x}}$) is: {mean:.2f}\n")

# Step 2: Calculate the squared difference for each value from the mean.
# We create a new list to store these calculated differences.
squared_differences = []
for value in values:
    # Calculate the difference between the value and the mean.
    difference = value - mean
    # Square the difference.
    squared_difference = difference ** 2
    # Add the result to our list.
    squared_differences.append(squared_difference)

print("Individual squared differences:")
for i in range(len(values)):
    print(f"Value {values[i]}: ({values[i]} - {mean:.2f})^2 = {squared_differences[i]:.2f}")

# Step 3: Sum all the squared differences to get the final result.
sum_of_squared_differences = sum(squared_differences)

print(f"\nThe sum of the squared differences ($\Sigma(x - \bar{{x}})^2$) is: {sum_of_squared_differences:.2f}")
```

#### 10. Opening, reading and writing files

```python
"""
This program demonstrates how to open, write to, and read from a text file in Python.
It uses the 'with' statement, which is the recommended way to handle files,
as it ensures the file is automatically closed after the block is executed.
"""

def write_and_read_file(filename, content):
    """
    Writes content to a file and then reads it back.

    Args:
        filename (str): The name of the file to create and use.
        content (str): The text content to write to the file.
    """
    # --- Writing to a file ---
    print(f"--- Writing content to '{filename}' ---")
    try:
        # Open the file in 'write' mode ('w'). If the file doesn't exist,
        # it will be created. If it exists, its content will be overwritten.
        with open(filename, 'w') as file:
            file.write(content)
        print("Successfully wrote content to the file.")
    except IOError as e:
        print(f"An error occurred while writing to the file: {e}")
        return  # Exit the function if writing fails

    print("\n--- Reading content from the file ---")
    # --- Reading from a file ---
    try:
        # Open the file in 'read' mode ('r'). This is the default mode.
        with open(filename, 'r') as file:
            # Read all the content from the file into a variable.
            read_content = file.read()
            print("Successfully read content from the file.")
            print("--- File Content ---")
            print(read_content)
            print("--------------------")
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
    except IOError as e:
        print(f"An error occurred while reading the file: {e}")

if __name__ == "__main__":
    # Define the name of the file to be created
    my_filename = "my_text_file.txt"

    # Define the content to be written to the file
    file_content = """This is the first line of my text file.
This is the second line.
And this is the third and final line.
"""
    # Call the function to perform the operations
    write_and_read_file(my_filename, file_content)
```

#### 11. Create a function the will simulate rolling two die and return the value of each and and the sum 

```python
"""
This program demonstrates how to open, write to, and read from a text file in Python.
It also includes a function to simulate rolling two dice.
"""

import random


def roll_dice():
    """
    Simulates rolling two six-sided dice and returns the value of each die and their sum.

    Returns:
        tuple: A tuple containing (die1_value, die2_value, total_sum).
    """
    # Simulate a roll of a six-sided die
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    total = die1 + die2
    return die1, die2, total

if __name__ == "__main__":


    # --- Demonstrate the new roll_dice() function ---
    print("\n--- Simulating a dice roll ---")
    die1_roll, die2_roll, total_sum = roll_dice()
    print(f"You rolled a {die1_roll} and a {die2_roll}.")
    print(f"The total sum is: {total_sum}")
    print("-----------------------------")
```

#### 12. Write the code needed to evaluate whether a value in a group of values is negative or postive and prints an approriate message, i.e. postive, negative 
-1.59, 2.36, 1.69, 1.13, -0.91, 1.48, -0.34, 1.31, -0.74, 0.2

```python
"""
This program evaluates a group of values to determine if each is positive or negative.
It iterates through a list of numbers and prints an appropriate message for each one.
"""

def check_values(values):
    """
    Iterates through a list of numbers and determines if each is positive, negative, or zero.

    Args:
        values (list): A list of numbers (floats or integers).
    """
    print("--- Evaluating values ---")
    for value in values:
        if value < 0:
            print(f"The value {value} is negative.")
        elif value > 0:
            print(f"The value {value} is positive.")
        else:
            print(f"The value {value} is zero.")
    print("-------------------------")

if __name__ == "__main__":
    # The group of values to be evaluated
    data_points = [-1.59, 2.36, 1.69, 1.13, -0.91, 1.48, -0.34, 1.31, -0.74, 0.2]

    # Call the function to check the values
    check_values(data_points)
```

#### 14.  Complete the following:
1. Write a program that generates 500 random integers from 1 to n, for example, the first integer will be 1 the second will be 1 or 2, the fifth between 1 and 5, and so on.
2. The program should write each integer to a file 
3. Once all integers are written close the file

##### Part 2
1. Open and read the file into a list
2. Create a line plot of the data

```python
"""
This program generates a series of random integers, writes them to a file,
and then reads the file to create a line plot of the data.

The program requires the `matplotlib` library. If you don't have it installed,
you can install it using pip: `pip install matplotlib`
"""
import random
import matplotlib.pyplot as plt

def generate_and_write_data(filename, count=500):
    """
    Generates a specified number of random integers and writes them to a file.
    The i-th integer is generated between 1 and i+1.

    Args:
        filename (str): The name of the file to write the integers to.
        count (int): The total number of integers to generate.
    """
    print(f"--- Generating {count} random integers and writing to '{filename}' ---")
    try:
        # Open the file in write mode ('w').
        with open(filename, 'w') as file:
            for i in range(1, count + 1):
                # Generate a random integer between 1 and the current iteration number (i).
                rand_num = random.randint(1, i)
                # Write the integer followed by a newline character.
                file.write(str(rand_num) + '\n')
        print("Successfully generated and wrote all integers to the file.")
    except IOError as e:
        print(f"An error occurred while writing to the file: {e}")

def read_data_and_plot(filename):
    """
    Reads integers from a file and creates a line plot.

    Args:
        filename (str): The name of the file to read from.
    """
    print(f"\n--- Reading data from '{filename}' and creating a plot ---")
    data_points = []
    try:
        # Open the file in read mode ('r').
        with open(filename, 'r') as file:
            for line in file:
                # Convert each line (a string) to an integer and add to the list.
                data_points.append(int(line.strip()))
        print("Successfully read data from the file.")
        
        # Create a line plot of the data.
        plt.figure(figsize=(12, 6))
        plt.plot(data_points, marker='o', markersize=2)
        
        # Add titles and labels for clarity.
        plt.title('Randomly Generated Integers (1 to n)', fontsize=16)
        plt.xlabel('Integer Number', fontsize=12)
        plt.ylabel('Value', fontsize=12)
        plt.grid(True)
        
        # Display the plot.
        print("Displaying the line plot. Close the plot window to continue.")
        plt.show()

    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found. Please ensure it was created successfully.")
    except (IOError, ValueError) as e:
        print(f"An error occurred while reading or processing the file: {e}")

if __name__ == "__main__":
    data_filename = "random_numbers.txt"
    
    # Part 1: Generate and write the data.
    generate_and_write_data(data_filename)
    
    # Part 2: Read the data and create the plot.
    read_data_and_plot(data_filename)
    
    print("\nProgram finished.")
```
